```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as c
import astropy.units as u
%matplotlib inline
```

```python
phi = np.arange(0,2*np.pi,.01)
r_orbit = 1.5e11 * u.m # change this for orbital radius
v_orbit = 3.e4 *u.m/u.s # change this for orbital speed
l = r_orbit*v_orbit
const = l**2/(c.G*c.M_sun)
e = 0.5
r = const/(1+e*np.cos(phi))
```

```python
ax = plt.subplot(111, projection='polar')
ax.plot(phi, r)
#ax.set_rticks([0.5,1,1.5,2])
```
