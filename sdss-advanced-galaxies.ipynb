```python
from astropy.io import ascii
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = ascii.read('sdss-galaxies-abell-2255.csv',delimiter=',')
```

## Plotting Positions ##

First I am going to plot the positions of the galaxies on the sky, just to see where they are located.

```python
plt.plot(data['ra'],data['dec'],'bo')
plt.xlabel('RA (deg)',fontsize=20)
plt.ylabel('Dec (deg)',fontsize=20)
```

## Plot Color-Magnitude Diagram ##

Now I will create another plot showing the color and magnitude of the galaxies.

For those of you new to magnitudes:
* larger magnitudes correspond to fainter galaxies, so a galaxy with $r = 12$ is brighter than a galaxy with $r = 15$

* if you subtract two magnitudes, you get a color.

* typically, astronomers subtract a bluer magnitude from a redder magnitude, like $g-r$ is green minus red

* when you subtract two magnitudes, a lower value indicates a bluer object, and a larger color value indicates redder objects (again, this assumes you are subtracting a shorter wavelength filter from a larger wavelength filter).


```python
plt.figure()
plt.plot(data['r'],data['g']-data['r'],'bo')
plt.xlabel('r (mag)',fontsize=20)
plt.ylabel('g-r (mag)',fontsize=20)
```

I have a really wacky point at $g-r > 4$, so I am going to reset my y limits to ignore that one... 

```python
plt.figure()
plt.plot(data['r'],data['g']-data['r'],'bo')
plt.xlabel('r (mag)',fontsize=20)
plt.ylabel('g-r (mag)',fontsize=20)
plt.ylim(.3,1.2)
```

Note $r >20$ is getting pretty faint, so we shouldn't pay too much attention to those points.
